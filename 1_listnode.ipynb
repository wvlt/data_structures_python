```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
kitchen = ListNode(5)
bedroom = ListNode(10)
garage = ListNode(20)
```

```python
kitchen.next = bedroom
bedroom.next = garage
```

```python
kitchen.val
```

```python
kitchen.next
```

```python
first = ListNode(5)
second = ListNode(10)
```

```python
print(first.val)
```

```python
print(first.next)
```

```python
print(second.val)
```

```python
print(second.next)
```

```python
first.next = second
```

```python
print(first.next)
```

```python
current = first
```

```python
print(current.val)
```

```python
current = second
print(current.val)
```

```python
current = current.next
```

```python
print(current)
```
